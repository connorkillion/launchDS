```
# Guess the Number
import random
start_range = input ("Choose a range of numbers. Low Number: ")
end_range = input ("High Number: ")
count = 5
number = random.randint(int(start_range) + 1, int(end_range))
print("Ok! Guess the number between " + str(start_range) + " and " + str(end_range) + " that I am thinking of. You get 5 tries.")
guess = int(input ("Guess: "))
while (guess != number):
  if (guess < number):
    print("Too low!")
    count -= 1
  elif (guess > number):
    print("Too high!")
    count -= 1
  if (count == 0):
    print("You're out of tries. The correct answer was " + str(number) + ". You lose.")
    break
  guess = int(input ("Guess: "))

if (guess == number):
  print("You win!")
  

```
